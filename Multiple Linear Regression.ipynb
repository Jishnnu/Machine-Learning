# LINEAR REGRESSION PROJECT 01 : MULTIPLE LINEAR REGRESSION

```python
# IMPORT ALL THE NECESSARY LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"/Multiple Linear Regression/Car_Models.csv")
data
```

```python
car_info = pd.get_dummies(data.car)
car_info
```

```python
# CLEANING THE DATA

cleaned_data = pd.concat([data, car_info], axis = 'columns')
cleaned_data = cleaned_data.drop(['car', 'Audi A5'], axis = 'columns')
cleaned_data
```

```python
# DETERMINING X & Y

regression = LinearRegression()
X = cleaned_data.drop(['sell_price'], axis = 'columns')
Y = cleaned_data.sell_price
```

```python
# FITTING THE MODEL

regression.fit(X,Y)
```

```python
# SCORE

regression.score(X, Y) * 100
```

```python
# PLOTTING THE GRAPH

plt.scatter(x=cleaned_data.mileage, y=cleaned_data.sell_price, color='blue', marker='+')
plt.scatter(x=cleaned_data.age, y=cleaned_data.sell_price, color='magenta', marker='+')
```
