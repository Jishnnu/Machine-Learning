# LOGISTIC REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
reg = linear_model.LogisticRegression()
```

```python
df = pd.read_csv(r"/Machine Learning/HR.csv")
df
```

```python
dummies = pd.get_dummies(df.salary)
dummies
```

```python
merged = pd.concat([df,dummies])
merged
```

```python
semifinal = merged.drop(['salary'],axis='columns')
semifinal
```

```python
import math
final = semifinal.fillna({
    'satisfaction_level': df.satisfaction_level.fillna(math.floor(df.satisfaction_level.median())),
    'last_evaluation': df.last_evaluation.fillna(math.floor(df.last_evaluation.median())),
    'number_project': df.number_project.fillna(math.floor(df.number_project.median())),
    'average_montly_hours': df.average_montly_hours.fillna(math.floor(df.average_montly_hours.median())),
    'time_spend_company': df.time_spend_company.fillna(math.floor(df.time_spend_company.median())),
    'Work_accident': df.Work_accident.fillna(math.floor(df.Work_accident.median())),
    'left': df.left.fillna(math.floor(df.left.median())),
    'promotion_last_5years': df.promotion_last_5years.fillna(math.floor(df.promotion_last_5years.median())),
    'Department':'No Department',
    'high': 0,
    'low': 0,
    'medium': 0
})
```

```python
x = final[['satisfaction_level','average_montly_hours','promotion_last_5years','high','low','medium']]
x
```

```python
y = final.left
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.predict(x_test)
```

# SUPPORT VECTOR MACHINE (SVM)

```python
from sklearn.datasets import load_digits
from sklearn.svm import SVC
digits = load_digits()
dir(digits)
```

```python
df = pd.DataFrame(digits.data,digits.target)
df['target'] = digits.target
df
```

```python
x_train,x_test,y_train,y_test= train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size=0.3)
```

### USING RBF

```python
rbf_model = SVC(kernel='rbf')
rbf_model.fit(x_train,y_train)
rbf_model.score(x_test,y_test)
```

### USING LINEAR

```python
rbf_model = SVC(kernel='linear')
rbf_model.fit(x_train,y_train)
rbf_model.score(x_test,y_test)
```

# RANDOM FOREST

```python
from sklearn.metrics import confusion_matrix
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data)
df['target'] = iris.target
df
```

```python
x_train,x_test,y_train,y_test= train_test_split(df.drop(['target'],axis='columns'),iris.target,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
confusionMatrix = confusion_matrix(y_test,y_predicted)
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(confusionMatrix,annot=True)
plt.xlabel('PREDICTED')
plt.ylabel('TRUTH')
```

# KFOLD CROSS VALIDATION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
dir(iris)
```

```python
df
```

```python
logisiticRegressionScore = cross_val_score(LogisticRegression(),iris.data,iris.target)
np.average(logisiticRegressionScore)
```

```python
SVCScore = cross_val_score(SVC(),iris.data,iris.target)
np.average(SVCScore)
```

```python
randomeForestScore = cross_val_score(RandomForestClassifier(n_estimators=80),iris.data,iris.target)
np.average(randomeForestScore)
```

```python
decisionTreeScore = cross_val_score(DecisionTreeClassifier(),iris.data,iris.target)
np.average(decisionTreeScore)
```
